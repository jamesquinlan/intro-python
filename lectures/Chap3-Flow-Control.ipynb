## Branching

```python
x = 1
y = 3

if x < y:
  print('Go to the school.')
else:
  print('Stay home, play Call of Duty.')
```

__Example__: Given a test score between 0 and 100%, convert to a letter grade based on standard, A $\ge 90$, B $\ge 80$, etc.

```python
grade = 67

# Determine Letter Grade
if grade >= 90:
  print('A')
elif grade >=80:
  print('B')
elif grade >=70:
  print('C')
elif grade >= 60:
  print('D')
else:
  print('F')
```

```python
# Why is this code better (assuming interger values)
if grade > 89:
  print('A')
elif grade > 79:
  print('B')
elif grade > 69:
  print('C')
elif grade > 59:
  print('D')
else:
  print('F')
```

# Repitition Structures
"_What computers are best at_."

* For loop
* while loop

```python
# Repitition Structure
# for loop and while loop

# while loop is best used if an unknown number of repititions is needs.

# For loop is best when there are a specific number of items to iterate over.


# Lazy Affirmations
how_many = 2

for i in range(how_many):
  print('I will make lots of money.')
```

### Note

It is better to loop over the items than the index, then get the items.  For example(s):

```python
A = [4, 1, 9, 8]

for i in range(len(A)):
  print(A[i])

print('-----------------')

for i in A:
  print(i)
```

## Alternative

```python
s = range(20)
for i in  s[::3]:
  print(i)

for i in range(0,20,3):
  print(i)

# equivalently as below
for i in range(20)[::3]:
  print(i)

for i in range(20)[2::3]:
  print(i)


```
