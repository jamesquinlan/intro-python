### IBM Watson and Cognitive Computing

* IBM Watson (or IBM's Watson) is a cloud-based cognitive-computing platform being employed across a wide range of real-world applications.

* Cognitive-computing systems simulate pattern-recognition and decision-making of the human brain (over time with the addition of more data).



* Watson was trained using machine-learning and reinforced learning techniques.

The Watson **Software Development Kit** (SDK) provides programmatic access to Watson services.  The range of capabilities allow the user to create **mashups** by applying serveral of the services provided by Watson.  These services are dynamic and are evolving.
